```python
def CreerDossierSauvegarde(ou,doss):
    nomEmplacementSauvegarde = ou + "/" + doss
    if not os.path.exists(nomEmplacementSauvegarde):
        os.mkdir(nomEmplacementSauvegarde)
        return nomEmplacementSauvegarde
    else:
        return nomEmplacementSauvegarde
```

```python
def finish(name):
    print(f" :) done.")

def DownloadVideo(linkVideo,ouCa,i = None):
    v = YouTube(linkVideo)
    print("... ",v.title)
    name = f"{v.title}" # ajouter fonctionnalité : ne pas telecharger si ya déjà un bail
    v.streams.get_highest_resolution().download(output_path=ouCa,filename=(name))
    v.register_on_complete_callback(finish(name))

def DownloadPlaylist(link):
    listplay = Playlist(link)
    Where = CreerDossierSauvegarde(CreerDossierSauvegarde(DRIVE_LEDOSSIER,'SAVES'),listplay.title)
    print(f"/////////// {listplay.title} : GO Download this Playlist ! ")
    for linkVideo in listplay:
        DownloadVideo(linkVideo,Where)
```
